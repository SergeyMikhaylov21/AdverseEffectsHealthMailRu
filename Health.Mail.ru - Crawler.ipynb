```python
# -*- encoding: utf-8 -*-
import lxml.html, requests, time, re, os
from lxml.cssselect import CSSSelector
from fake_useragent import UserAgent
```

```python
ua = UserAgent()
headers = {'User-Agent': ua.random}
```

```python
#в mainurl записаны результаты поиска по слову "побочка"

mainurl = 'https://health.mail.ru/search/consultations/?query=%D1%80%D0%B0%D1%81%D1%81%D0%B5%D1%8F%D0%BD%D0%BD%D1%8B%D0%B9+%D1%81%D0%BA%D0%BB%D0%B5%D1%80%D0%BE%D0%B7'
consults_refs = set()

def refs_hunter(url):
    r = requests.get(url, headers=headers)
    tree = lxml.html.fromstring(r.text)
    sel_refs = CSSSelector('a[href]')
    refs = sel_refs(tree)
    page_refs = [page_ref.get('href') for page_ref in refs]
    for r in page_refs:
        consult = re.findall('\/consultation\/[0-9]+\/', r)
        consults_refs.update(consult)
```

```python
refs_hunter(mainurl)
for it in range(2, 55):
    next_url = 'https://health.mail.ru/search/consultations/?query=%D1%80%D0%B0%D1%81%D1%81%D0%B5%D1%8F%D0%BD%D0%BD%D1%8B%D0%B9+%D1%81%D0%BA%D0%BB%D0%B5%D1%80%D0%BE%D0%B7&page=' + str(it)
    refs_hunter(next_url)
```

```python
for i in consults_refs:
    print(i)
```

```python
os.mkdir('запрос рассеянный склероз')

i = 1
for cons in consults_refs:
    try:
        url = 'https://health.mail.ru' + str(cons)
        r = requests.get(url, headers=headers)
        tree = lxml.html.fromstring(r.text)
        sel_text = CSSSelector('div.entry__description')
        tree_text = sel_text(tree)
        txt = [div.text for div in tree_text if not div.text.endswith('...')]
        with open('запрос рассеянный склероз/' + str(i) + '.txt', 'a', encoding='utf-8') as new_txt:
            for t in txt:
                new_txt.write(t + '\n')
        i += 1
    except:
        continue
```
